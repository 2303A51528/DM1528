```python
import pandas as pd

# Load dataset (replace with your file path or URL)
# The original URL is giving a 404 error.
# Using an updated URL from Kaggle for the Titanic dataset.
url = 'https://storage.googleapis.com/tf-datasets/titanic/train.csv'
data = pd.read_csv(url)
```

```python
# Generate statistical summary of the dataset
summary = data.describe()
```

```python
# Display the summary
print(summary)
```

```python
# Summary of categorical columns (mode, unique values, etc.)
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
# Plot histogram for 'Age' column
# Check if the column name is 'age' instead of 'Age'
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'Pclass' (Passenger class)
plt.figure(figsize=(6, 4))
```

```python
plt.figure(figsize=(6, 4))
# Check if 'Pclass' column exists, if not, try variations
if 'Pclass' in data.columns:
    data['Pclass'].value_counts().plot(kind='bar', color='lightcoral')
plt.figure(figsize=(6, 4))
# Check if 'Pclass' column exists, if not, try variations
if 'Pclass' in data.columns:
    data['Pclass'].value_counts().plot(kind='bar', color='lightcoral')
elif 'pclass' in data.columns:
    data['pclass'].value_counts().plot(kind='bar', color='lightcoral')  # Try 'pclass'
else:
    pass
```

```python
# Box plot for 'Age' to visualize outliers and distribution
plt.figure(figsize=(6, 4))
# Change 'Age' to 'age' to match the actual column name
plt.boxplot(data['age'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns

# Pairplot to visualize relationships between numerical features
# Correct the column names to match the actual names in the dataset
# Based on the dataset, it's likely 'age' instead of 'Age', etc.
# Changed 'sibsp' to 'SibSp' to match the actual column name in the dataset.
sns.pairplot(data[['age', 'fare', 'parch']].dropna())
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data[['age', 'fare', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
# Correlation heatmap
correlation_matrix = data[['age', 'fare', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
# Count Plot
```

```python
# Count plot for 'Survived' (how many survived vs how many did not)
sns.countplot(x='survived', data=data, palette='Set2',legend=False)
plt.title('Survival Count')
plt.xlabel('Survived')
# Count plot for 'Survived' (how many survived vs how many did not)
sns.countplot(x='survived', data=data, palette='Set2',legend=False)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
# Violin Plot
```
